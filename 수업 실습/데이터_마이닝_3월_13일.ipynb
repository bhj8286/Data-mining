```python
# python의dictionary 자료형(key: value)을Series data로만들수있다.
sdata={'SongGain': 35000, 'HongJa': 67000, 'Girl': 12000, 'BTS': 4000}
# dictionary의key가Series의index가된다.
obj3 = pd.Series(sdata)
obj3
```

```python
# series 데이터인obj3의Series의이름을'Salary'로설정
obj3.name = 'Salary'
# series 데이터인obj3의인덱스(index)의이름을Names로함
obj3.index.name = "성명"
obj3
```

인덱스를 새로 지정하면 앞에서 지정해준 obj3.index.name = "성명"은 없어지는 것인가?

```python
#index  'SongGain', 'HongJa', 'Girl', ‘BTS’ ➔A,B,C,D로변경
obj3.index = ['A', 'B', 'C', 'D']
obj3
```
